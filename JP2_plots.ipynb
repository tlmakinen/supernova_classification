```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# cosmo param plot

a = np.linspace(0.0001, 1, num=300)

z = np.linspace(500, 0)

def omegam(a, omega_mo):
    raw = (omega_mo / (a**3))    
    return raw

def omegarad(a, omega_ro):
    raw = omega_ro / (a**4)
    return raw 

def omega_de(a, omega_lamb, w):
    eps = (3/2)*(1 + w)
    raw = omega_lamb / (a**(2*eps))
    return raw

def E_a(a, w, omega_mo, omega_ro, omega_lamb):
    raw =  omegam(a, omega_mo) + 0.7 + omegarad(a, omega_ro)
    return raw
```

```python
# parameters
omega_mo = 0.3
omega_ro = 9.28E-5
omega_lamb = 0.7
w_pars = [-1, -1.1, -0.9]
w_labels = [r'$\Omega_{\Lambda}$, $w$ = -1.0', r'$\Omega_{DE}$, $w$ = -1.1', r'$\Omega_{DE}$, $w$ = -0.9']
```

```python
w = -1.0
plt.plot(np.log10(a), omegam(a, omega_mo) / E_a(a, w, omega_mo, omega_ro, omega_lamb))
```

```python
fig,ax = plt.subplots(1, 1, figsize=(15,9))
omegaM, = ax.semilogx((a), omegam(a, omega_mo) / E_a(a, -1.0, omega_mo, omega_ro, omega_lamb), linewidth=3.0)
omegaRad, = ax.semilogx(a, omegarad(a, omega_ro) / E_a(a, -1.0, omega_mo, omega_ro, omega_lamb), linewidth=3.0)

for w in range(len(w_pars)):
    plt.semilogx(a, omega_de(a, omega_lamb, w_pars[w]) / E_a(a, w, omega_mo, omega_ro, omega_lamb), linewidth=3.0, label=w_labels[w])
    
ax.set_ylim(0, 1.)
ax.set_xlim(10**(-4), 1)

ax.set_xlabel(r'$log ( \frac{a}{a_o} )$', fontsize = 30)
ax.set_ylabel('$\Omega_i$', fontsize = 26)

plt.xticks(fontsize = 22)
plt.yticks(fontsize = 22)

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.9, box.height])

# Put a legend to the right of the current axis
#ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1, fancybox=True, shadow=True, fontsize=26)
ax.legend(loc='best', ncol=1, fancybox=True, shadow=True, fontsize=26)

# Create a legend for the first line.
omegaRad_label = plt.text(x=8**(-3), y=0.2, s = r'$\Omega_{rad}$', fontsize = 26)
plt.text(x=7.5**(-3), y = 0.8, s=r'$\Omega_m$', fontsize=26)

# Add the legend manually to the current Axes.
#ax = plt.gca().add_artist(omegaRad_label)

plt.savefig('/c/Users/lucas/Documents/Fall2018/Imperial_2018/JP2/plots/paper_figs/fraction_evolution.png', dpi='figure')

plt.show()
```
